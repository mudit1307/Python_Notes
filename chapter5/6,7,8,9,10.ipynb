```python
#9

#list vs strings 

#strings are immutable 
#lists are mutable 

s="string"
s.title()
print(s)
t=s.title()
print(t)

l=['word1','word2','word3']
l.pop()
print(l)
l.append('word3')
print(l)

#looping in list 
fruits=['orange','apple','pear','kiwi']
#for loop in list 
for  i in fruits:
    print(i)
    
#while loop in list 
i=0 
while i<len(fruits):
    print(fruits[i])
    i+=1
```
