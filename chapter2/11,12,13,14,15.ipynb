```python
#11

# solve problem with spaces using strip method 
name="   Harshit    "
dots='.....'

print(name+dots)
print(name.lstrip()+dots)     #lstrip()method 
print(name.rstrip()+dots)     #rstrip()method 
print(name.strip()+dots)      #strip()method 

name1 ='   hars  hit   '
print(name1.replace(" ",'')+dots)    #replace()method 


#EXAMPLE (IT ALSO WORK WITH STRING )
print("  ya sh  ".replace(" ",""))
```
