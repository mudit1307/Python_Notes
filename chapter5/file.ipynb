```python
#exercise1
'''
define a function which will take list containing numbers as input 
and return list containing square of every elements 

example 
numbers=[1,2,3,4]
square_list(numbers) ----> return ---->[1,4,9,16]
'''

def square(l):
    # squ=[]
    for i in range(len(l)):
        l[i]=l[i]*l[i]
        # squ.append(i*i)
    return l
    
numbers=list(range(1,11))
print(numbers)
print(square(numbers))


print('\n')
#or 
def square_list(l):
    squ=[]
    for i in range(len(l)):
        squ.append(i**2)
    return l
    
numbers=list(range(1,11))
print(numbers)
print(square(numbers))

```

```python
#exercise3
'''define a function that takes list of words as argument and 
return list with reverse of every element in that list 

example 
['abc','tuv','xyz'] --->['cba','vut','zyx']
'''

def rev_item(l):
    revit=[]
    for i in l:
        revit.append(i[::-1])
    return revit    

words=['abc','def','name']
print(rev_item(words))
```

```python
#exercise4 
''' filter odd even 
define a function 
input
list ------> [1,2,3,4,5,6,7]

output ----->[[1,3,5,7],[2,4,6]]
'''

def eve_odd(l):
    eve=[]
    odd=[]
    final=[]
    for i in l:
        if i%2==0:
            eve.append(i)
        else:
            odd.append(i)
    final.append(odd)
    final.append(eve)
    return final 
    
num=[1,2,3,4,5,6,7]
print(eve_odd(num))


#or 
def eve_odd2(l):
    eve=[]
    odd=[]
    
    for i in l:
        if i%2==0:
            eve.append(i)
        else:
            odd.append(i)
    l.clear()
    l.append(odd)
    l.append(eve)
    
    return l
    
    
nu=[1,2,3,4,5,6,7]

print(eve_odd2(nu))

#or 
def filter_odd_even(l):
    odd=[]
    even=[]
    for i in l:
        if i%2==0:
            even.append(i)
        else:
            odd.append(i)
    output=[odd,even]
    return output
    
num=list(range(1,8))
print('\n')
print(filter_odd_even(num))



```

```python

#exercise5
'''common element finder function 
define a function which takes 2 lists as a input and return a list 
which contains common elements of both lists 

example
input ----->[1,2,5,8],[1,2,7,6]
output ----->[1,2]
'''

def  common_element(a,b):
    output=[]
    for i in a:
        if i in b:
            output.append(i)
        else:
            pass
    return output 

num1=[1,2,5,8]
num2=[1,2,7,6]

print(common_element(num1,num2))
```

```python
#min and max function 
numbers=[6,60,2]
print(min(numbers))
print(max(numbers))

#example
def greatest_diff(l):
    a=max(l)
    b=min(l)
    return a-b
print(greatest_diff(numbers))
```

```python
#exercise 6
''' function counts number of lists inside the input list 
[1,2,3,[2,3],[2,4]],--->input 
2--->output 

'''
def  no_list(l):
    out=0
    for i in l:
        if type(i)== list:
            out+=1
    return out 

number=[1,2,3,[12,3],123,[23,23,3]]
print(no_list(number))

```
