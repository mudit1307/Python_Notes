```python
with open('file7.html','r') as webpage:
    with open('output8.txt','w') as wf:
        for line in webpage.readlines():
            if '<a href=' in line:
                first_quote=line.find("\"")
                second_quote=line.find("\"",first_quote+1)
                url=line[first_quote+1:second_quote]
                wf.write(url+ "\n")
```

```python
# to read long .txt file in parts 
# .read(no. of character to read)

with open('long10.txt','r')as f:
    data = f.read(1000)
    print(data)
```

```python
#to read complete file in parts (one by one)

with open('long10.txt','r')as f:
    data = f.read(1000)
    print(data)
    while len(data) > 0:
        print(data)
        data = f.read(1000)
```
