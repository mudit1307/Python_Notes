```python
#add and delete data from dictionaries 

#add data in deictionaries 
user_info={
    "name":"harshit",
    'age': 24,
    'movies':['coco','kimi'],
    'tunes':['awake','fairy']
}

user_info['fav_songs']=['song1','song2']
print(user_info)

#.pop()method

popped_items=user_info.pop('tunes')
print("popped items is {}".format(popped_items))
print(user_info)
print(type(popped_items))

popped= user_info.pop('age')
print(type(popped))

#.popitem()method 

popped_item=user_info.popitem()
print(user_info)
print(popped_item)
```

```python
#exercise 1 

'''
define a that takes a number(n) 
return a dictionary containing cube of numbers from 1 to numbers

#example 
cube_finder(3)
{1:1,2:8,3:27}
'''

def cube(n):
    d={}
    for i in range(1,n+1):
        d[i]=i**3
    
    return d


print(cube(5))


#exercise 2
'''
ask user for user_information then add it in dictionary
and print the dictionary in diff. lines using loop 
'''


user={}

name=input("what is your name: ") 
age=input("what is your age: ")
fav_movies=input('your fav movies seperated by commas: ').split(",")
fav_tunes=input('your fav tunes seperated by comma: ').split(",")

user['name']=name
user['age']=age 
user['fav_movies']=fav_movies
user['fav_tunes']=fav_tunes
print(user)

for key,value in user.items():
    print("{}:{}".format(key ,value))

```
