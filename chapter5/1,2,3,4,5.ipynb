```python
#1

#CHAPTER 5

#INTRO TO LIST 
#IT IS AN DATA STRUCTURE 
#LIST IS AN ORDERED COLLECTIOIN OF ITEMS.
#YOU CAN STORE SNYTHING IN LIST INT,STRING,FLOAT.
#WE CAN CREATE OUR LIST USING SQUARER BRACKETS.


#DECLARING LISTS 
numbers=[1,2,3,4]
print(numbers)

words=['word1','word2','word3']
print(words)

mixed=[1,2,3,'four','five','six',2.3,None]
print(mixed)


#accessing the elements

numbers=[1,2,3,4]
print(numbers[1])
print(numbers[:1])
print(numbers[-1])
```

```python
#2


#changing in stored data in list 

numbers=[1,2,3,4]

numbers[1]='two'
print(numbers)

numbers=[1,2,3,4]
numbers[1:]="two"   #here string is taken as a complete list.
print(numbers)

num=[1,2,3,4]
num[1:]=['two','three','four']         
print(num)
mixed=[1,2,3,4,'five']
mixed[1:]=['two','three','four']           #string length will decresses 
print(mixed)



```

```python
#3 


# add data to list 
#.append() method 
fruits=["apple",'orange']
fruits.append("mango")
print(fruits)

fut=[]
fut.append("orange")
fut.append("apple")
print(fut)

# more method to add data 
#.insert() method 

fruts=['apple','orange']
fruts.insert(1,'grapes')
print(fruts)
fruts.insert(10,'mango')
print(fruts)


```

```python
#4


#to concatenate two or more 

fruits1=['apple','orange']
fruits2=['mango']
fruits3=['guava']
frt=fruits1+fruits2+fruits3
print(frt)

#.extend() method

frt1=['mango','apple']
frt2=['orange']
frt1.extend(frt2)

print(frt1)
print(frt2)


#difference between append and extend method  
frt1=['mango','apple']   #it willl add list into list
frt2=['orange']
frt1.append(frt2)

print(frt1)



```

```python
#5 


#delete data from list 


#.pop()method 
#use to delete specific single item from the list 

fruits=['orange','apple','pear','banana','kiwi']
fruits.pop()

print(fruits)
fruits.pop(1)
print(fruits)


#del operator 
#can use to del sud list using string slicing 
fruits=['orange','apple','pear','banana','kiwi']
del fruits[2]
print(fruits)

fruits=['orange','apple','pear','banana','kiwi']
del fruits[:2]
print(fruits)


#.remove()method
#use to delete item whose position is not known 
fruits=['orange','apple','pear','banana','kiwi']
fruits.remove('banana')
# fruits.remove('mango')  #valueError
print(fruits)

fruits=['orange','apple','pear','apple','banana','kiwi']
fruits.remove('apple')    #in case of same items it will delete first one of them.
print(fruits)
```
